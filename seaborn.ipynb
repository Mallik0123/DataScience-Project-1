```python
#SEABORN

import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
print(df.to_string())
g=sb.FacetGrid(df,col="time")#facetget is used to do subplot
g.map(plt.hist,"tip")#map is used to plot
plt.show()
```

```python
g=sb.FacetGrid(df,col="time",hue="smoker")#hue is used for different colours
g.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
g=sb.FacetGrid(df,col="sex",hue="smoker")
g.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
#Statistical estimation
sb.barplot(x="sex",y="tip",hue="time",data=df)
plt.show()
```

```python
#Linear realtionship
sb.regplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
#pair plot
sb.pairplot(df) #it take related column with each other and diagonal and diagonal graghs are unigraph
plt.show()
```

```python
#pairplot seaborn:plotting selected variables
selected_vars=['total_bill','tip']
sb.pairplot(df,vars=selected_vars)
plt.show()
```
